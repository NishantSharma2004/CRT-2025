```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Creating the dataset
data = {
    'Student Name': ['John', 'Sarah', 'David', 'Emily'],
    'Math': [80, 95, 70, 85],
    'Science': [75, 85, 65, 90],
    'English': [90, 80, 75, 88],
    'History': [70, 85, 80, 92]
}

df = pd.DataFrame(data)

```

```python
# **23. Calculate the mean score of each subject using NumPy**
subject_means = df.iloc[:, 1:].mean().to_dict()
print("\nMean Score of Each Subject:\n", subject_means)

```

```python
# **24. Calculate the average score of each student**
df["Average Score"] = df.iloc[:, 1:].mean(axis=1)
print("\nAverage Score of Each Student:\n", df[['Student Name', 'Average Score']])

```

```python
# **25. Find the student with the highest overall score**
df["Total Score"] = df.iloc[:, 1:-1].sum(axis=1)
top_student = df.loc[df["Total Score"].idxmax(), "Student Name"]
print(f"\nStudent with the Highest Overall Score: {top_student}")

```

```python
# **26. Scatter plot for Math vs Science Scores**
plt.figure(figsize=(6, 4))
plt.scatter(df["Math"], df["Science"], color='blue')
plt.xlabel("Math Scores")
plt.ylabel("Science Scores")
plt.title("Math vs Science Scores")
plt.grid()
plt.show()

```

```python
# **27. Standardize the scores using NumPy**
standardized_scores = (df.iloc[:, 1:-2] - df.iloc[:, 1:-2].mean()) / df.iloc[:, 1:-2].std()
print("\nStandardized Scores:\n", standardized_scores)

```

```python
# **28. Visualize the total scores using a bar plot**
plt.figure(figsize=(6, 4))
plt.bar(df["Student Name"], df["Total Score"], color=['red', 'green', 'blue', 'purple'])
plt.xlabel("Students")
plt.ylabel("Total Score")
plt.title("Total Scores of Students")
plt.xticks(rotation=45)
plt.show()

```

```python
# **29. Pie chart representing subject-wise average scores**
plt.figure(figsize=(6, 6))
plt.pie(subject_means.values(), labels=subject_means.keys(), autopct='%1.1f%%', colors=['yellow', 'blue', 'red', 'green'])
plt.title("Average Scores by Subject")
plt.show()

```

```python
# **30. Display students whose History score is greater than the average History score**
average_history_score = df["History"].mean()
high_history_students = df[df["History"] > average_history_score]
print("\nStudents with History Score Greater than the Average:\n", high_history_students[['Student Name', 'History']])
```
